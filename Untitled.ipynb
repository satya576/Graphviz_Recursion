# Reversort

```python
def reverSort(A):
    cost  = 0
    n = len(A)
    for i in range(n-1):
        subarray = A[i:n]
        j = i + subarray.index(min(A[i:n]))
        A[i:j+1] = A[i:j+1][::-1]
        cost = cost + (j - i +1)
    return cost
```

```python
A = [2,5,4,8, 1]
cost = reverSort(A)
print(A)
print(cost)
```

## make list from given cost and length of the array

```python
def makeList(n, c, m):
    if n == 1:
        return [m]
    else:
        if c-1 >= n -2 and c-1 <= n*(n-1)/2 -1 :
            return [m] + makeList(n-1, c - 1, m+1)
        else:
            ans = makeList(n-1, c-n, m+1)
            ans = ans[::-1] + [m]
            return ans
```

```python
n = 7
c = n*(n+1)/2 - 1
if c >= n-1 and c <= n*(n+1)/2 - 1:
    A = makeList(n, c, 1)
    print(A)
    print(reverSort(A))
else:
    print("NOT POSSIBLE")
```

## Permutations

```python
def permute(A, DS, flag, ans):
    if len(DS) == len(A):
        ans.append(DS[:])
        return
    for i in range(len(A)):
        if flag[i] != True:
            DS.append(A[i])
            flag[i] = True
            permute(A, DS, flag, ans)
            DS.pop()
            flag[i] = False
```

```python
A = [1,2,3]
DS = []
flag = [False, False, False]
ans = []
permute(A, DS, flag, ans)
print(ans)
print(len(ans))
```

## permutations with repetations

```python
def permute(A, DS, ans):
    if len(DS) == len(A):
        ans.append(DS[:])
        return
    for i in range(len(A)):
        DS.append(A[i])
        permute(A, DS, ans)
        DS.pop()
```

```python
A = [1,2,3]
DS = []
ans = []
permute(A, DS, ans)
print(ans)
print(len(ans))
```

## permutations with repetations of length $r$ from the array of length $n$

```python
def permute(A, r, DS, ans):
    if len(DS) == r:
        ans.append(DS[:])
        return
    for i in range(len(A)):
        DS.append(A[i])
        permute(A, r,  DS, ans)
        DS.pop()
```

```python
A = [1,2,3,4]
DS = []
ans = []
r = 2
permute(A, r, DS, ans)
print(ans)
print(len(ans))
```
